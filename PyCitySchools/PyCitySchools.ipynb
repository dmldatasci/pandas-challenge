# School District Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending 645-675 per student actually underperformed compared to schools with smaller budgets (585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
import pandas as pd
import os

schools_file_path = os.path.join('Resources', 'schools_complete.csv')
schools_frame = pd.read_csv(schools_file_path)

students_file_path = os.path.join('Resources', 'students_complete.csv')
students_frame = pd.read_csv(students_file_path)
```

## District Summary

```python
# extract total number of unique schools
school_count = len(schools_frame['School ID'].unique())
school_count
```

```python
# extract the total number of unique students
student_count = len(students_frame['Student ID'])
student_count
```

```python
# extract the total budget for the district
budget = schools_frame['budget'].sum()
budget
```

```python
# calculate the average math score
avg_math_score = students_frame['math_score'].mean()
avg_math_score
```

```python
# calculate the average math score
avg_read_score = students_frame['reading_score'].mean()
avg_read_score
```

```python
# calculate percent of students who passed math
math_pass_percentage = len(students_frame[students_frame['math_score'] >= 70]) / student_count * 100
math_pass_percentage
```

```python
# calculate percentage of students who passed reading
read_pass_percentage = len(students_frame[students_frame['reading_score'] >= 70]) / student_count * 100
read_pass_percentage
```

```python
# calculate percentage of students who passed both math & reading
mr_pass_percentage = len(students_frame[(students_frame['math_score'] >= 70) & (students_frame['reading_score'] >= 70)]) / student_count * 100
mr_pass_percentage
```

```python
# construct an appropriate dataframe
district_summary = pd.DataFrame({
    'Total Schools' : [school_count],
    'Total Students' : [student_count],
    'Total Budget' : [budget],
    'Average Math Score' : [avg_math_score],
    'Average Reading Score' : [avg_read_score],
    '% Passing Math' : [math_pass_percentage],
    '% Passing Reading' : [read_pass_percentage],
    '% Overall Passing' : [mr_pass_percentage]})

# format relevant values from the dataframe
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

district_summary
```

## School Summary

```python
# construct a new dataframe with the information we need and to which we can add calculations
per_school_summary = schools_frame.rename(columns={
    'school_name' : 'School Name',
    'type' : 'School Type',
    'size' : 'Total Students',
    'budget' : 'Total School Budget'
    }).sort_values(by=['School Name']).set_index(['School Name'])

per_school_summary
```

```python
# `schools_frame` already contains school type
per_school_summary['School Type']
```

```python
# `schools_frame` already contains total students
per_school_summary['Total Students']
```

```python
# calculate per student budget
per_school_summary['Per Student Budget'] = per_school_summary['Total School Budget'] / per_school_summary['Total Students']
per_school_summary
```

```python
# calculate average math score
avg_scores = students_frame[['school_name', 'math_score', 'reading_score']].groupby(['school_name']).mean()
per_school_summary['Average Math Score'] = avg_scores['math_score']
per_school_summary['Average Reading Score'] = avg_scores['reading_score']
per_school_summary
```

```python
# calculate the percentage of students at each school who passed math
per_school_summary['% Passing Math'] = students_frame[students_frame['math_score'] >= 70].groupby(['school_name']).count()['math_score'] / per_school_summary['Total Students'] * 100

# calculate the percentage of students at each school who passed reading
per_school_summary['% Passing Reading'] = students_frame[students_frame['reading_score'] >= 70].groupby(['school_name']).count()['reading_score'] / per_school_summary['Total Students'] * 100

# calculate the percentage of students at each school who passed both math and reading
per_school_summary['% Passing Overall Passing'] = students_frame[(students_frame['reading_score'] >= 70) & (students_frame['math_score'] >= 70)].groupby(['school_name']).count()['reading_score'] / per_school_summary['Total Students'] * 100

per_school_summary
```

```python
# edit formatting for more readable output
per_school_summary['Total Students'] = per_school_summary['Total Students'].astype('int32').map("{:,}".format)
per_school_summary['Total School Budget'] = per_school_summary['Total School Budget'].map("${:,.2f}".format)
per_school_summary['Per Student Budget'] = per_school_summary['Per Student Budget'].map("${:,.2f}".format)

per_school_summary
```
